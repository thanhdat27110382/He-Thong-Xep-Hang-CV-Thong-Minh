```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
```

```python
# Khởi tạo trình duyệt Chrome
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized") 
options.add_argument("--disable-blink-features=AutomationControlled")  
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)
```

```python
driver.get("https://zety.com/cv-examples")
time.sleep(5)  # Wait for initial page load
print(driver.page_source)
```

```python
# Hàm cuộn trang để tải toàn bộ nội dung
def scroll_to_bottom():
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Chờ để trang tải thêm nội dung
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
```

```python
# Cuộn trang để đảm bảo tải hết dữ liệu
scroll_to_bottom()
```

```python
# Khởi tạo danh sách để lưu dữ liệu
data = []
```

```python
# Tìm tất cả các div có class="box"
box_elements = driver.find_elements(By.CLASS_NAME, "box")

# Duyệt qua từng box để lấy dữ liệu
for box in box_elements:
    # Tìm ul với class="box__list mt-16--phone" trong mỗi box
    try:
        ul_element = box.find_element(By.CLASS_NAME, "box__list.mt-16--phone")
        # Tìm tất cả các li trong ul
        li_elements = ul_element.find_elements(By.CLASS_NAME, "box__list-item")
        
        # Duyệt qua từng li để lấy href và text của thẻ a
        for li in li_elements:
            try:
                a_tag = li.find_element(By.TAG_NAME, "a")
                href = a_tag.get_attribute("href")
                category = a_tag.text.strip()
                
                # Thêm dữ liệu vào danh sách
                data.append({
                    "resume_link": href,
                    "Category": category
                })
            except Exception as e:
                print(f"Lỗi khi lấy dữ liệu từ thẻ a: {e}")
    except Exception as e:
        print(f"Lỗi khi tìm ul trong box: {e}")
```

```python
print(data)
```

```python
# Đóng trình duyệt
driver.quit()
```

```python
# Kiểm tra dữ liệu trước khi lưu
if data:
    # Chuyển dữ liệu thành DataFrame và lưu vào file CSV
    df = pd.DataFrame(data)
    df.to_csv("D:/BaiDoAnChuyenNganh3/Automated-Resume-Ranking-System-main/csvfiles/crawlcv/crawldata2links_examples.csv", index=False, encoding="utf-8")
    print(f"Dữ liệu đã được crawl và lưu vào file cv_examples.csv với {len(data)} mục")
else:
    print("Không có dữ liệu nào được crawl!")
```
