```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Khởi tạo trình duyệt Chrome
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized") 
options.add_argument("--disable-blink-features=AutomationControlled")  
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)
```

```python
driver.get("https://resumekraft.com/cv-examples/")
time.sleep(5)  # Wait for initial page load
print(driver.page_source)
```

```python
# Hàm cuộn trang đến cuối để tải hết nội dung
def scroll_to_bottom():
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Chờ 2 giây để tải nội dung mới
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:  # Nếu không còn nội dung mới, thoát vòng lặp
            break
        last_height = new_height

# Cuộn trang đến cuối
print("Đang cuộn trang để tải hết nội dung...")
scroll_to_bottom()
print("Đã tải hết nội dung trang.")
```

```python
# Danh sách để lưu dữ liệu
data = []
```

```python
try:
    # Tìm section
    all_industry_section = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.ID, "section-11"))
    )

    # Tìm div class "catWrap"
    container = all_industry_section.find_element(By.CLASS_NAME, "catWrap")

    # Tìm div class "catSec"
    blade_content = container.find_element(By.CLASS_NAME, "catSec")

    # Tìm div class "catTxt"
    cat_txt = blade_content.find_element(By.CLASS_NAME, "catTxt")

    # Tìm thẻ ul trong catTxt
    ul_tag = cat_txt.find_element(By.TAG_NAME, "ul")

    # Tìm tất cả các thẻ li trong ul
    li_tags = ul_tag.find_elements(By.TAG_NAME, "li")

    print(f"Tìm thấy {len(li_tags)} phần tử li.")

    # Duyệt qua từng li
    for li in li_tags:
        try:
            # Tìm thẻ a trong li
            a_tag = li.find_element(By.TAG_NAME, "a")
            
            # Lấy link từ href
            resume_link = a_tag.get_attribute("href")
            
            # Lấy nội dung của thẻ a
            category = a_tag.text.strip()
            
            # Lọc bỏ các thẻ a không chứa thông tin ngành nghề (như "See all ...")
            if "See all" in category:
                continue
            
            # Thêm vào danh sách dữ liệu
            data.append({
                "Category": category,
                "Resume_link": resume_link
            })
        except Exception as e:
            print(f"Lỗi khi xử lý thẻ li: {e}")
            continue

except Exception as e:
    print(f"Lỗi khi truy cập cấu trúc HTML: {e}")
```

```python
# Đóng trình duyệt
driver.quit()
```

```python
# Chuyển dữ liệu thành DataFrame
df = pd.DataFrame(data)
```

```python
# Xóa phần "Resume Example" trong cột Category
df['Category'] = df['Category'].str.replace(' Resume Example', '', regex=False)
df['Category'] = df['Category'].str.replace(' CV Example', '', regex=False)
df['Category'] = df['Category'].str.replace(' Resume Sample', '', regex=False)
df['Category'] = df['Category'].str.replace(' CV Sample', '', regex=False)
```
